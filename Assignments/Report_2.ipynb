```python
#Summarizing Sales by Year and Product Category
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/MSSP6070/Assignments/Week 7/InternetSales.csv', encoding='ISO-8859-1')
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
df['OrderYear'] = df['OrderDate'].dt.year
sales_by_year_product = df.groupby(['OrderYear', 'EnglishProductName'])['SalesAmount'].sum().reset_index()
sales_by_year_product.head()
```

```python
#Summary Statistics for Yearly Income and Gender by Country Region
summary_statistics = df.groupby('CountryRegionCode').agg(
    YearlyIncome_min=('YearlyIncome', 'min'),
    YearlyIncome_max=('YearlyIncome', 'max'),
    YearlyIncome_std=('YearlyIncome', 'std'),
    Gender_count=('Gender', 'count')
).reset_index()
summary_statistics.head()
```

```python
#Average Price of Product Category with Most Sales
product_sales = df.groupby('EnglishProductName').agg(
    TotalSalesAmount=('SalesAmount', 'sum'),
    AverageUnitPrice=('UnitPrice', 'mean')
).reset_index()
top_product = product_sales.sort_values(by='TotalSalesAmount', ascending=False).iloc[0]
top_product[['EnglishProductName', 'TotalSalesAmount', 'AverageUnitPrice']]
```

```python
#Impact of Homeownership on Sales
sales_by_homeownership = df.groupby('HouseOwnerFlag')['SalesAmount'].sum().reset_index()
sales_by_homeownership.head()
```
