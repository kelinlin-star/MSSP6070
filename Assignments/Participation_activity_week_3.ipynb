```python
def collatz(number):
    if number % 2 == 0:
        result = number // 2
    else:
        result = 3 * number + 1
    print(result)
    return result


def main():
    print("Enter a positive integer:")
    try:
        n = int(input())
        if n < 1:
            print("Please enter a number greater than 0.")
            return
        while n != 1:
            n = collatz(n)
    except ValueError:
        print("Invalid input. Please enter an integer.")

main()
```
