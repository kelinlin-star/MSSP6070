```
def collatz(number: int) -> int:
    """
    Calculates the next number in the Collatz sequence based on the rules.

    Parameters:
        number (int): The current positive integer.

    Returns:
        int: The next number in the Collatz sequence.
    """
    if number % 2 == 0:
        # If the number is even
        return number // 2
    else:
        # If the number is odd
        return 3 * number + 1


def main():
    """
    The main function, handles user input and prints the Collatz sequence.
    """
    # Loop until the user provides a valid positive integer
    while True:
        try:
            print("Please enter a positive integer (greater than 0):")
            # Get user input
            n_input = input()
            # Try to convert the input to an integer
            num = int(n_input)

            if num < 1:
                print("⚠️ Invalid input: Please ensure the number is greater than 0.")
                # Continue the loop to ask for re-entry
                continue

            # Valid input received, break the loop
            break

        except ValueError:
            print(f"❌ Invalid input: '{n_input}' is not a valid integer.")
            # Continue the loop to ask for re-entry

    # Print the Collatz sequence
    print(f"\nCollatz sequence starting from {num}:")
    print(num, end=' ') # Print the starting number

    # Loop until the number reaches 1
    while num != 1:
        # Call the collatz function to calculate the next number
        num = collatz(num)
        # Print the result, separated by a space
        print(num, end=' ')

    print("\nSequence finished.")

# Standard Python practice to ensure main() runs only when the script is executed directly
if __name__ == "__main__":
    main()
```
