```python
aapl = pd.read_csv("data/aapl.csv")
amzn = pd.read_csv("data/amzn.csv")
nflx = pd.read_csv("data/nflx.csv")
fb = pd.read_csv("data/fb.csv")
goog = pd.read_csv("data/goog.csv")
```

```python
aapl['ticker'] = 'AAPL'
amzn['ticker'] = 'AMZN'
nflx['ticker'] = 'NFLX'
fb['ticker']   = 'FB'
goog['ticker'] = 'GOOG'
```

```python
faanf = pd.concat([aapl, amzn, nflx, fb, goog])
faanf.to_csv('WeeklyModules/Week05/faanf.csv', index=False)
```

```python
faanf = pd.read_csv('WeeklyModules/Week05/faanf.csv')
faanf['date'] = pd.to_datetime(faanf['date'])
faanf['volume'] = faanf['volume'].astype(int)
faanf = faanf.sort_values(['date', 'ticker'])
```

```python
faanf.nlargest(7, 'volume').reset_index(drop=True)
```

```python
order = ['open', 'high', 'low', 'close', 'volume']

melted = pd.melt(
    faanf,
    id_vars=['date', 'ticker'],
    value_vars=order,
    var_name='variable',
    value_name='value'
)

melted['variable'] = pd.Categorical(melted['variable'], categories=order, ordered=True)
melted = melted.sort_values(['date', 'ticker', 'variable']).reset_index(drop=True)

pd.set_option('display.float_format', '{:.2f}'.format)

print(melted.head(25))
```
