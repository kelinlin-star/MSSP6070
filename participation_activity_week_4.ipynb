```python
folder_path = '/content/drive/MyDrive/MSSP6070/Assignments/Week 4 assignments/Working with Pandas DataFrames'
file_path = '/content/drive/MyDrive/MSSP6070/Assignments/Week 4 assignments/Working with Pandas DataFrames/parsed.csv'
import pandas as pd
data = pd.read_csv(file_path)
```

```python
#a. Find the 95th percentile of earthquake magnitude in Japan using magType of 'mb'
japan_earthquakes = data[data['parsed_place'] == 'Japan']
mb_earthquakes = japan_earthquakes[japan_earthquakes['magType'] == 'mb']

percentile_95 = mb_earthquakes['mag'].quantile(0.95)
percentile_95
```

```python
#b. Find the percentage of earthquakes in Indonesia that were coupled with tsunamis
indonesia_earthquakes = data[data['parsed_place'] == 'Indonesia']
tsunami_percentage = (indonesia_earthquakes['tsunami'].sum() / len(indonesia_earthquakes)) * 100
tsunami_percentage
```

```python
#c. Get summary statistics for earthquakes in Nevada
nevada_earthquakes = data[data['parsed_place'].str.contains('Nevada')]
nevada_summary = nevada_earthquakes.describe()
nevada_summary
```

```python
#d. Add a column to the dataframe indicating whether or not the earthquake happened in a country or US state that is on the Ring of Fire
#Create a list of Ring of Fire locations.
ring_of_fire_locations = ['Alaska', 'California', 'Indonesia', ' Nevada', 'Mexico']

#Add a new column to the dataframe indicating if the earthquake occurred in one of these locations.
data['in_ring_of_fire'] = data['parsed_place'].apply(lambda x: x in ring_of_fire_locations)

data.head()
```

```python
#e. Calculate the number of earthquakes in the Ring of Fire locations and the number outside of them
earthquakes_in_ring_of_fire = data[data['in_ring_of_fire']].shape[0]
earthquakes_outside_ring_of_fire = data[~data['in_ring_of_fire']].shape[0]
print(f"Number of earthquakes in the Ring of Fire: {earthquakes_in_ring_of_fire}")
print(f"Number of earthquakes outside of the Ring of Fire: {earthquakes_outside_ring_of_fire}")
```

```python
#f. Find the tsunami count along the Ring of Fir
tsunami_count_in_ring_of_fire = data[data['in_ring_of_fire']]['tsunami'].sum()
print(f"Tsunami count along the Ring of Fire: {tsunami_count_in_ring_of_fire}")
```
